```python
import re
from collections import defaultdict
import matplotlib.pyplot as plt

def roman_to_arabic(roman):
    
    roman_values = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50,
        'C': 100, 'D': 500, 'M': 1000
    }
    
    result = 0
    prev_value = 0
    for char in reversed(roman):
        curr_value = roman_values[char]
        if curr_value >= prev_value:
            result += curr_value
        else:
            result -= curr_value
        prev_value = curr_value
    return result

def parse_date(date_str):
    
    if not date_str or date_str.isspace():
        return None
        
    date_str = date_str.strip()
    
    if 'BC' in date_str and 'AD' in date_str:
        match = re.match(r'(\d+)\s*BC\s*-\s*(\d+)\s*AD', date_str)
        if match:
            start = -int(match.group(1))
            end = int(match.group(2))
            return (start + end) / 2
    
    if 'cent.' in date_str:
        centuries = date_str.replace(' cent.', '').split('-')
        if len(centuries) == 2:
            start_year = (roman_to_arabic(centuries[0]) - 1) * 100 + 1
            end_year = roman_to_arabic(centuries[1]) * 100
            return (start_year + end_year) / 2
        else:
            century = roman_to_arabic(centuries[0])
            return ((century - 1) * 100 + 1 + century * 100) / 2
    
    
    if 'BC' in date_str:
        years = date_str.replace(' BC', '').split('-')
        if len(years) == 2:
            return -(abs(int(years[0])) + abs(int(years[1]))) / 2
        else:
            return -abs(int(years[0]))
    
    if '-' in date_str:
        years = date_str.split('-')
        if years[1] == '':  
            start_year = int(years[0])
            return start_year + 50  
        elif years[0] and years[1]:
            return (int(years[0]) + int(years[1])) / 2
    
    if date_str.isdigit():
        return int(date_str)
        
    return None

def analyze_dates(dates):
    
    years = [parse_date(date) for date in dates if date.strip()]
    years = [year for year in years if year is not None]
    
    min_year = min(years)
    max_year = max(years)
    span = max_year - min_year
    num_bins = 200  
    
    plt.figure(figsize=(15, 6))
    plt.hist(years, bins=num_bins, edgecolor='black')
    plt.title('Distribution of Historical Years')
    plt.xlabel('Year')
    plt.ylabel('Count')
    
    plt.grid(True, linestyle='--', alpha=0.7)
    
    
    plt.xticks(rotation=45)
    
    print(f"Earliest Year: {min_year:.0f}")
    print(f"Latest Yea: {max_year:.0f}")
    print(f"Year Span: {span:.0f}")
    
    return years


years = analyze_dates(df['year'])
plt.show()
```

```python
import pandas as pd
import re
from typing import List, Dict, Tuple
from collections import defaultdict

import requests
from time import sleep

class WikidataAPI:

    def __init__(self):
        self.base_url = "https://www.wikidata.org/w/api.php"
        self.cache = {}
    
    def get_entity_id(self, label: str) -> str:
        """获取实体的Wikidata ID"""
        if label in self.cache:
            return self.cache[label]
            
        params = {
            "action": "wbsearchentities",
            "format": "json",
            "language": "en",
            "search": label
        }
        
        try:
            response = requests.get(self.base_url, params=params)
            data = response.json()
            
            if "search" in data and data["search"]:
                entity_id = data["search"][0]["id"]
                self.cache[label] = entity_id
                return entity_id
            
            return None
        except Exception as e:
            print(f"Error fetching Wikidata ID for {label}: {str(e)}")
            return None
        finally:
            sleep(0.5)  

class LocationProcessor:
    def __init__(self, raw_data: str):
        self.wikidata_api = WikidataAPI()
        self.raw_data = raw_data
        self.processed_locations = []
        
    def clean_data(self) -> List[str]:
        """clean raw data and return unique locations"""
        
        unique_locations = sorted(list(set(self.raw_data)))
        return unique_locations
    
    def parse_location(self, location: str) -> Dict:
        """parse location into city, region, country"""
        result = {
            'label': location,
            'wikidata_id': None,
            'type': None
        }
        
        # process alternative names
        if ' / ' in location:
            names = location.split(' / ')
            main_name = names[0]
            location = main_name
            result['alternative_names'] = names[1:]
        
        # separate city, region, country
        if ', ' in location:
            parts = location.split(', ')
            city = parts[0]
            country = parts[-1]
            
            # process region if available
            if len(parts) > 2:
                region = parts[1]
                result.update({
                    'city': city,
                    'region': region,
                    'country': country,
                    'type': 'city'
                })
            else:
                result.update({
                    'city': city,
                    'country': country,
                    'type': 'city'
                })
        else:
            # only country
            result.update({
                'country': location,
                'type': 'country'
            })
        
        # get Wikidata ID
        try:
            wikidata_id = self.wikidata_api.get_entity_id(location)
            if wikidata_id:
                result['wikidata_id'] = wikidata_id
        except Exception as e:
            print(f"Error getting Wikidata ID for {location}: {str(e)}")
        
        return result
        
    
    def process_all(self) -> List[Dict]:
        """process all locations"""
        unique_locations = self.clean_data()
        processed_data = []
        
        for location in unique_locations:
            processed = self.parse_location(location)
            print(processed)
            processed_data.append(processed)
            
        self.processed_locations = processed_data
        return processed_data
    
    def get_statistics(self) -> Dict:
        """"""
        if not self.processed_locations:
            self.process_all()
            
        stats = {
            'total_locations': len(self.processed_locations),
            'countries': set(),
            'cities': 0,
            'regions': 0
        }
        
        for loc in self.processed_locations:
            if loc['type'] == 'country':
                stats['countries'].add(loc['country'])
            elif loc['type'] == 'city':
                stats['cities'] += 1
                stats['countries'].add(loc['country'])
                if 'region' in loc:
                    stats['regions'] += 1
                    
        stats['countries'] = len(stats['countries'])
        return stats


    #

processor = LocationProcessor(L)

processed_data = processor.process_all()

# get statistics
stats = processor.get_statistics()

print("\nProcessed Data Sample (first 5 entries):")
for entry in processed_data[:5]:
    print(entry)
    
print("\nStatistics:")
for key, value in stats.items():
    print(f"{key}: {value}")
    
```
